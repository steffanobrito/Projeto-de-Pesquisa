```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
sns.set_style("whitegrid")
```

Leitura de dados

```python
dfga= pd.read_excel("OldplanilhagrupoA.xlsx")
```

Analise dos dados

```python
dfga
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(dfga.drop('Target',axis=1))
```

```python
scaled_features = scaler.transform(dfga.drop('Target',axis=1))
```

```python
scaled_features
```

```python
df_f = pd.DataFrame(scaled_features,columns=dfga.columns[:-1])
df_f.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_features,dfga['Target'],shuffle=False,
                                                    test_size=0.33)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

```python
error_rate = []


for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Taxa de Erro vs. Valor de K')
plt.xlabel('K')
plt.ylabel('Taxa de Erro')
plt.savefig('Oldescolha de k p grupo 1')
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=10)

knn.fit(X_train,y_train)
pred = knn.predict(X_test)

print('WITH K=10')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```

```python
from sklearn.model_selection import cross_val_score
valid=cross_val_score(knn,df_f,dfga['Target'],cv=6)
print(valid)
```

```python
plt.figure(figsize=(14,8))
plt.plot(valid,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Precisão(Cross Validation) x Número de Iterações')
plt.xlabel('CV-Núm. de Iterações')
plt.ylabel('Taxa de Precisão(Cross Validation)')
plt.axis((-0.1,5.05,0,1.05))
plt.savefig('OLdcross validation grupo 1 c 6i')
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
valid10=cross_val_score(knn,df_f,dfga['Target'],cv=10)
print(valid10)
```

```python
plt.figure(figsize=(14,8))
plt.plot(valid10,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Precisão(Cross Validation) x Número de Iterações')
plt.xlabel('CV-Núm. de Iterações')
plt.ylabel('Taxa de Precisão(Cross Validation)')
plt.axis((-0.1,9.10,0,1.01))
plt.savefig('Oldcross validation grupo 1 c 10i')
plt.show()
```

```python
from sklearn import metrics
fpr,tpr,thresholds = metrics.roc_curve(y_test,pred,pos_label=1)
roc_auc = metrics.auc(fpr, tpr)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,pred)
```

```python
plt.figure(figsize=(14,8))
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([-0.01, 1.01])
plt.ylim([-0.01, 1.01])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
#plt.savefig('OldRoc p grupo 1')
plt.show()
```
